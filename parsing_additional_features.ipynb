```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
from tqdm import tqdm, tqdm_notebook
from tqdm import tqdm, tqdm_notebook
tqdm.pandas()
```

```python
PATH = 'med cards/1 рота/'
```

```python
cards = [f for f in listdir(PATH) if isfile(join(PATH, f))]
```

```python
df = pd.read_excel(PATH+cards[0])
```

```python
def get_age(df):
    data = df[['Unnamed: 55', 'Unnamed: 57']].loc[5]
    age = float(data.iloc[0])*10 + float(data.iloc[1])
    return age
    
def get_nationality(df):
    nationality = df['Unnamed: 1'].loc[7] 
    return nationality

def get_date_of_birth(df):
    day = str(df['Unnamed: 37'].loc[5]) + str(df['Unnamed: 37'].loc[5])
    month = str(df['Unnamed: 41'].loc[5]) + str(df['Unnamed: 43'].loc[5])
    year = str(df['Unnamed: 45'].loc[5]) + str(df['Unnamed: 47'].loc[5])
    date = f'{day}.{month}.{year}'
    return date
    

def get_place_of_birth(df):
    city = df['Unnamed: 1'].loc[9] 
    return city

def previous_place_of_living(df):
    city = df['Unnamed: 1'].loc[11] 
    return city

def get_bite_type(df):
    bite = df['Unnamed: 33'].loc[11] 
    return bite
    
def get_fluorosis(df):
    fluorosis = df['Unnamed: 49'].loc[45]
    return fluorosis

def get_sum_hygiene_index(df):
    h_index =  df['Unnamed: 45'].loc[93]
    return(h_index)

print(get_age(df))
print(get_nationality(df))
print(get_place_of_birth(df))
print(previous_place_of_living(df))
print(get_bite_type(df))
print(get_fluorosis(df))
print(get_sum_hygiene_index(df))
print(get_date_of_birth(df))
```

```python
data = pd.read_csv('stoma_dataset.csv')
```

```python
def calc_new_features(idx, function):
    number = idx[0]
    PATH = f'med cards/{number} рота/'
    df = pd.read_excel(PATH+idx)
    return function(df)
```

```python
data['age'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_age))
```

```python
data['nationality'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_nationality))
data['date_of_birth'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_date_of_birth))
data['place_of_birth'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_place_of_birth))
data['previous_place_of_living'] = data['id'].progress_apply(lambda a: calc_new_features(a, previous_place_of_living))
data['bite_type'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_bite_type))
data['fluorosis'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_fluorosis))
data['sum_hygiene_index'] = data['id'].progress_apply(lambda a: calc_new_features(a, get_sum_hygiene_index))


```

```python
data.to_csv('extended_stoma_dataset.csv')
```
